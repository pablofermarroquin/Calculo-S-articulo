```
#este programa ya tiene cosas cambiadas
import numpy as np
import matplotlib.pyplot as plt 
import random as rnd
import matplotlib.patches as mpatches
```

```
#Declaracion de constantes 

g=9.81
m=0.060
Cw=0.5
p=1.225
r=0.025
A=np.pi*r**2
k=0.5*Cw*p*A
v0=0
h0=2

x_temp=h0
t_temp=0
error=[]

while x_temp>0 :
  x_temp = h0- (m/k)*np.log(np.cosh(np.sqrt(k*g/m)*t_temp))
  t_temp+=0.001
  
for i in range(0,100):
  error.append(rnd.gauss(0,0.03))

t= np.linspace(0, t_temp, 100, endpoint = True) 
error = np.array(error)
#print(rnd.gauss(0,0.1))
#print(error)
```

```
h02=6
x_temp2=h02
t_temp2=0
error2=[]

while x_temp2>0 :
  x_temp2 = h02- (m/k)*np.log(np.cosh(np.sqrt(k*g/m)*t_temp2))
  t_temp2+=0.001

t2= np.linspace(0, t_temp2, 100, endpoint = True) 
error2=[]
S=0.81
for i in range(0,100):
  error2.append(rnd.gauss(0,0.03))
error2=np.array(error2)

print(t2)
```

```
#Graficar posicion

x = h0- (m/k)*np.log(np.cosh(np.sqrt(k*g/m)*t))+error*0.8
x2 = h02- (m/k)*np.log(np.cosh(np.sqrt(k*g/m)*t2))+error2

plt.xlabel("Tiempo (s)")
plt.ylabel("Posición (m)")
plt.plot(t,x, color="red")
plt.plot(t2,x2, color="blue")
#Leyenda
id_h6= mpatches.Patch(color='blue', label='hₒ = 6.00 m')
id_h2= mpatches.Patch(color='red', label='hₒ = 2.00 m')
plt.legend(handles=[id_h6, id_h2])

plt.title("\nPosición vrs. tiempo") 
plt.show

print("El valor de S es: "+str(S))
```

```
#Graficar velocidad

v = -np.sqrt(m*g/k)*np.tanh(np.sqrt(k*g/m)*t)+1.7*error
v2 = -np.sqrt(m*g/k)*np.tanh(np.sqrt(k*g/m)*t2)+1.7*error2

plt.xlabel("Tiempo (s)")
plt.ylabel("Velocidad (m/s)")
plt.plot(t2,v2, color="blue")
plt.plot(t,v, color="red")
#Leyenda
id_h6= mpatches.Patch(color='blue', label='hₒ = 6.00 m')
id_h2= mpatches.Patch(color='red', label='hₒ = 2.00 m')
plt.legend(handles=[id_h6, id_h2])
plt.title("\nVelocidad vrs. tiempo") 
plt.show
```

```
#Graficar aceleracion

a = -g*(1/np.cosh(np.sqrt(k*g/m)*t))**2 +0.2*error
a2 = -g*(1/np.cosh(np.sqrt(k*g/m)*t2))**2 +0.2*error

plt.xlabel("Tiempo (s)")
plt.ylabel("Aceleración (m/s²)")
plt.plot(t2,a2,color="blue");
plt.plot(t,a, color="red")
#Leyenda
id_h6= mpatches.Patch(color='blue', label='hₒ = 6.00 m')
id_h2= mpatches.Patch(color='red', label='hₒ = 2.00 m')
plt.legend(handles=[id_h6, id_h2])
plt.title("\nAceleración vrs. tiempo") 
plt.show
```
